# Anomoly detection in CPU usage

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('./hardware_monitor_data.csv')
```

```python
df = df.drop(['cpu_temperature','cpu_power'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split

train_data, val_data = train_test_split(df['cpu_usage'].values, test_size=0.2, random_state=42)
train_data = train_data.reshape(-1, 1)
val_data = val_data.reshape(-1, 1)
```

```python
lof = LocalOutlierFactor(n_neighbors=20)
sample = train_data[1000:2000]
# Assuming `train_data` is 1D or has one feature
labels = lof.fit_predict(sample)  # -1 for anomalies, 1 for normal points

# Separate normal points and anomalies
normal_data = sample[labels == 1]
anomalies = sample[labels == -1]

# Generate x-axis indices for the data points
normal_indices = np.where(labels == 1)[0]
anomaly_indices = np.where(labels == -1)[0]

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(sample)), sample, c='blue', label='Validation Data', alpha=0.6)
plt.scatter(anomaly_indices, anomalies, c='red', label='Anomalies', alpha=0.9)
plt.title("Validation Data with Detected Anomalies (Isolation Forest)")
plt.xlabel("Index")
plt.ylabel("CPU Usage")
plt.legend()
plt.grid(True)
plt.show()
```
